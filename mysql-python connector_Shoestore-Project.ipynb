### Add a new customer to `customers` table

```python
def add_customer(customer_firstname, customer_lastname, customer_phonenum, customer_addr, cust_purchased_date, cust_qty, shoe_id):
    try:
        connection = mysql.connector.connect(
            host='localhost',
            database='shoestore',
            user='root',
            password='root'
        )
        if connection.is_connected():
            cursor = connection.cursor()
            insert_query = '''
                INSERT INTO customers (cust_fname, cust_lname, cust_phone, cust_addr, cust_purchased_date, cust_qty, sh_id)
                VALUES (%s, %s, %s, %s, %s, %s, %s);
            '''
            customer_data = (customer_firstname, customer_lastname, customer_phonenum, customer_addr, cust_purchased_date, cust_qty, shoe_id)
            cursor.execute(insert_query, customer_data)
            connection.commit()
            print(f"‚úÖ Customer {customer_firstname} {customer_lastname} added successfully")

    except Error as e:
        print("‚ùå Error:", e)




    
```

### Delete A Customer

```python
def delete_customer_account(customer_id):
    try:
        connection = mysql.connector.connect(
            host='localhost',
            database='shoestore',
            user='root',
            password='root'
        )
        if connection.is_connected():
            cursor = connection.cursor()
            cursor.execute("SELECT cust_fname, cust_lname FROM customers WHERE cust_id = %s;", (customer_id,))
            customer = cursor.fetchone()

            if not customer:
                print(f"‚ùå Customer with ID {customer_id} does not exist.")
                return
                
            delete_query = "DELETE FROM customers WHERE cust_id = %s;"
            cursor.execute(delete_query, (customer_id,))
            connection.commit()


            cust_fname, cust_lname = customer
            print(f"üóëÔ∏è Customer with ID {customer_id} named {cust_fname} {cust_lname} deleted successfully")

    except Error as e:
        print("‚ùå Error:", e)
```

### Add a Product to `shoes` Table

```python
def add_product(shoe_name, shoe_brand, shoe_price, shoe_available, employee_id):
    try:
        connection = mysql.connector.connect(
            host='localhost',
            database='shoestore',
            user='root',
            password='root'
        )
        if connection.is_connected():
            cursor = connection.cursor()
            insert_query = '''
                INSERT INTO shoes (sh_name, sh_brand, sh_price, sh_avail, emp_id)
                VALUES (%s, %s, %s, %s, %s);
            '''
            shoes_stock = (shoe_name, shoe_brand, shoe_price, shoe_available, employee_id)
            cursor.execute(insert_query, shoes_stock)
            connection.commit()
            print(f"‚úÖ {shoes_stock[3]} Product of '{shoes_stock[1]}' added successfully")

    except Error as e:
        print("‚ùå Error:", e)
```
